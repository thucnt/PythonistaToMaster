```python
import petl as etl
import pickle
from petl import appendpickle
```

```python
with open('./datasets/students.p', 'wb') as student:
    pickle.dump(['Name', 'Age', 'Gender', 'GPA'], student)
    pickle.dump(['Alan Adachi', 21, 'male', 3.67], student)
    pickle.dump(['Kathy Cruz', 24, 'female', 2.98], student)
    pickle.dump(['Kelly Davis', 23, 'female', 3.1], student)
    pickle.dump(['Fred Baker', 23, 'male', 2.67], student)
```

```python
open('./datasets/marketing-campaigns.xml', 'r').read()
```

```python
advertisement_data = etl.fromxml('./datasets/marketing-campaigns.xml', 'item',
                                {
                                    'campaign_name':'campaign_name',
                                    'start_date':'start_date',
                                    'end_date':'end_date',
                                    'budget':'budget',
                                    'roi':'roi',
                                    'type':'type',
                                    'target_audience':'target_audience',
                                    'channel':'channel',
                                    'conversion_rate':'conversion_rate',
                                    'revenue':'revenue'
                                })
```

```python
advertisement_data.display()
```

```python
lookup_name = etl.lookup(advertisement_data, ('channel', 'target_audience', 'type'), 'campaign_name')
```

```python
lookup_name[('promotion', 'B2C', 'social media')]
```

```python
lookup_one = etl.lookupone(advertisement_data, 'channel', 'type')
lookup_one
```

```python
lookup_channel = etl.dictlookup(advertisement_data, 'channel')
```

```python
lookup_channel['promotion']
```

```python
lookup_channel_type = etl.dictlookup(advertisement_data, ['channel', 'type'])
```

```python
lookup_channel_type[('promotion', 'social media')]
```

```python
advertisement_data.display(index_header=True)
```

```python
advertisement_data.display(tr_style=lambda row: 'background-color: %s' % ('#faa' if row.roi > '0.4' else 'white'))
```
